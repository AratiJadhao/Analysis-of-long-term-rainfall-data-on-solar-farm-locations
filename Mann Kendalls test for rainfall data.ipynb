```python
import numpy as np
import pandas as pd
import pymannkendall as mk
from scipy.stats import kendalltau
```

```python
df=pd.read_excel('India_Interpolated_Rainy_Days_RW.xlsx')
df1=pd.read_excel('d1.xlsx')
df2=pd.read_excel('d2.xlsx')
```

```python
del df['State']
del df['Area']
del df['Latitude']
del df['Longitude']
del df['fid']
del df['Altitude']
```

```python
del df[1988]
del df[1989]
del df[1990]
del df[1991]
del df[1992]
del df[1993]
del df[1994]
del df[1995]
del df[1996]
del df[1997]
del df[1998]
del df[1999]
```

```python
result=[]
for index, row in df.iterrows():
        tau, p_value = kendalltau(range(len(row)), row)
        result.append({ 'tau': tau, 'p_value': p_value,'trend': 'increasing' if tau > 0 else 'decreasing' if tau < 0 else 'no trend','trend_type':'Significant' if p_value<0.05 else 'Not significant'})
        
```

```python
result
```

```python
df3 = pd.DataFrame.from_dict(result)

print (df3)

df3.to_excel('result.xlsx')
```

```python

concatenated_file=pd.concat([df, df3], axis=1)
output_file_path = 'concatenated_file1.xlsx'
concatenated_file.to_excel(output_file_path, index=False)
concatenated_file.to_excel('India_Interpolated_Rainy_Days_RW1.xlsx')
```

```python
del concatenated_file['t_statistics']
del concatenated_file['p_values']
```
