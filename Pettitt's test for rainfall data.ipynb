```python
def pettitt_test(x):
    n = len(x)
    rank = rankdata(x)
    U = np.zeros(n)

    for k in range(n):
        U[k] = 2 * np.sum(rank[:k+1]) - (k+1) * (n+1)
    
    U = np.abs(U)
    K = np.argmax(U)
    U_max = U[K]
    p_value = 2 * np.exp((-6 * U_max**2) / (n**3 + n**2))
    return K, p_value
```

```python
change_points = []
p_values = []
for index, row in df.iterrows():
    data = row.values[1:]  
    K, p_value = pettitt_test(data)
    change_points.append(K)
    p_values.append(p_value)

# Add the results back to the DataFrame
df['Change_Point_Index'] = change_points
df['p_value'] = p_values

# Save the results back to the Excel file
df.to_excel('output_with_change_points.xlsx', index=False)

# Print the results for verification
print(df)
```
